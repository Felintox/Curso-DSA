## Pergunta de Negócio 1: <br>

## Qual Cidade com Maior Valor de Venda de Produtos da Categoria 'Office Supplies'?

```python
q1=df[df['Categoria']=='Office Supplies']
q1=q1.groupby(['Cidade']).sum()
q1=q1['Valor_Venda'].sort_values(ascending=False)
print(f'A cidade com maior Valor de Venda é {(q1.index[0]).upper()} com valor de {q1[0]}')
```

```python
q1=q1.reset_index()
q1=q1[q1['Valor_Venda']>=(q1['Valor_Venda'].quantile(0.99))]


plt.bar(q1['Cidade'],q1['Valor_Venda'],width=0.2)
plt.title('Cidades com Maiores Vendas')

for index, value in enumerate(q1['Valor_Venda']):
    plt.text(index, value, str(round(value,2)), ha='center', va='bottom')

plt.gca().axes.get_yaxis().set_visible(False)
for spine in plt.gca().spines.values():
    spine.set_visible(False)
    
plt.xticks(rotation=90)

plt.show()
```

#### Podemos verificar que a cidade com maior valor é "New York City" com um valor de "68362.81"

## Pergunta de Negócio 5:

### Qual Segmento Teve o Maior Total de Vendas?


```python
# Agrupamos por segmento e calculamos o total de vendas
q5= df.groupby('Segmento')['Valor_Venda'].sum().reset_index().sort_values(by ='Valor_Venda',
                                                                                      ascending = False)

#Tirando a notação cientifica que o numero estava! (1.133834e+06) -> 1133834.33
q5['Valor_Venda']=q5['Valor_Venda'].apply(lambda x: float(f'{x:.2f}'))
q5.head()
```

```python
# Gráfico de barras simples
plt.figure(figsize=(16, 6), dpi=300)

# Barras com cores diferentes
plt.bar(q5['Segmento'], q5['Valor_Venda'], color=['#2E86AB', '#A23B72', '#F18F01'])

# Valores nas barras com porcentagem
for i, value in enumerate(q5['Valor_Venda']):
    percentage = (value/sum(q5['Valor_Venda'])*100)
    value_formatted = f"{value:,.0f}".replace(",", ".")
    plt.text(i, value + 20000, f'${value_formatted}\n({percentage:.1f}%)', 
             ha='center', fontweight='bold')

# Labels
plt.title('Total de Vendas Por Segmento', fontsize=16, fontweight='bold')
plt.ylabel('Valor de Vendas ($)')

# Total no canto direito
total_formatted = f"{int(sum(q5['Valor_Venda'])):,}".replace(",", ".")
plt.text(0.98, 0.95, f'Total: ${total_formatted}', 
         transform=plt.gca().transAxes, ha='right', va='top',
         bbox=dict(boxstyle='round', facecolor='lightgray', alpha=0.8))

plt.show()
```

#### O segmento de "Consumer" tem 50,72% das vendas, seguidos por "Home Office" com 18,90% e "Corporate" com 30,38%

## Pergunta de Negócio 7:

- Se o Valor_Venda for maior que 1000 recebe 15% de desconto.
- Se o Valor_Venda for menor que 1000 recebe 10% de desconto.

### Quantas Vendas Receberiam 15% de Desconto?

```python
# Cria uma nova coluna de acordo com a regra definida acima
df['Desconto'] = np.where(df['Valor_Venda'] > 1000, 0.15, 0.10)
df
```

## Pergunta de Negócio 8 :

### Considere Que a Empresa Decida Conceder o Desconto de 15% do Item Anterior. Qual Seria a Média do Valor de Venda Antes e Depois do Desconto?

```python
Média_antes=df['Valor_Venda'].mean()
Média_antes
```

```python
#Se aplicamos apenas o 15%, precisamos retirar o desconto de 10% então vou deixar 1 para não multiplicarmos
df['Desconto15']=np.where(df['Desconto'] == 0.1,1,0.15)

df['Média_Depois']=df['Valor_Venda']-(df['Valor_Venda']*df['Desconto15'])
média_depois=df['Média_Depois'].mean()
```

```python
resultado=Média_antes-média_depois
print(resultado)
```

#### Nos daria uma diferença de 145.7 de média

# Pergunta de Negócio 9 

## Qual o Média de Vendas Por Segmento, Por Ano e Por Mês?

```python
# Extraímos o mês e gravamos em uma nova variável
df['Mes'] = df['Data_Pedido'].dt.month
df[['Mes','Valor_Venda']]
```

```python
# Agrupamos por ano, mês e segmento e calculamos estatísticas de agregação
q9= df.groupby(['Ano', 'Mes', 'Segmento'])['Valor_Venda'].agg([np.mean])
q9
```

```python
# Vamos extrair os níveis
anos = q9.index.get_level_values(0)
meses =q9.index.get_level_values(1)
segmentos = q9.index.get_level_values(2)

# Plot
plt.figure(figsize = (12, 6))
sns.set()
fig1 = sns.relplot(kind = 'line',
                   data = q9, 
                   y = 'mean', 
                   x = meses,
                   hue = segmentos, 
                   col = anos,
                   col_wrap = 4)
plt.show()
```

#### O gráfico acima representa a distribuição de cada segmento no decorrer dos meses dos anos de  2015 até 2018

# Pergunta de Negócio 10 : <br>

## Qual o Total de Vendas Por Categoria e SubCategoria, Considerando Somente as Top 12 SubCategorias? 



```python
# Agrupamos por categoria e subcategoria e calculamos a soma somente para variáveus numéricas
q10 = df.groupby(['Categoria',
                  'SubCategoria']).sum(numeric_only = True).sort_values('Valor_Venda',
                                                                        ascending = False).head(12)
```

```python
# Convertemos a coluna Valor_Venda em número inteiro e classificamos por categoria
q10=q10[['Valor_Venda']].astype(int).sort_values(by ='Categoria').reset_index()
# Dataframe com categorias e subcategorias
q10
```

Obs: Classificar o item acima por categoria é importante para preencher o gráfico com as subcategorias para cada categoria de forma ordenada.

```python
# Criamos outro dataframe somente com os totais por categoria
q10_cat = q10.groupby('Categoria').sum(numeric_only = True).reset_index()
# Dataframe com categorias 
q10_cat
```

```python
# Listas de cores para categorias
cores_categorias = ['#5d00de',
                    '#0ee84f',
                    '#e80e27']
# Listas de cores para subcategorias
cores_subcategorias = ['#aa8cd4',
                       '#aa8cd5',
                       '#aa8cd6',
                       '#aa8cd7',
                       '#26c957',
                       '#26c958',
                       '#26c959',
                       '#26c960',
                       '#e65e65',
                       '#e65e66',
                       '#e65e67',
                       '#e65e68']
```

```python
# Gráfico de barras por categoria
plt.figure(figsize=(12, 6), dpi=300)

# Barras com cores personalizadas
plt.bar(q10_cat['Categoria'], q10_cat['Valor_Venda'], color=['#5d00de', '#0ee84f', '#e80e27'])

# Valores nas barras com porcentagem
for i, value in enumerate(q10_cat['Valor_Venda']):
    percentage = (value/sum(q10_cat['Valor_Venda'])*100)
    value_formatted = f"{value:,.0f}".replace(",", ".")
    plt.text(i, value + 15000, f'${value_formatted}\n({percentage:.1f}%)', 
             ha='center', fontweight='bold')

# Labels
plt.title('Total de Vendas por Categoria', fontsize=16, fontweight='bold')
plt.ylabel('Valor de Vendas ($)')

# Total no canto direito
total_formatted = f"{int(sum(q10_cat['Valor_Venda'])):,}".replace(",", ".")
plt.text(0.98, 0.95, f'Total: ${total_formatted}', 
         transform=plt.gca().transAxes, ha='right', va='top',
         bbox=dict(boxstyle='round', facecolor='lightgray', alpha=0.8))

plt.show()
```

#### No Gráfico acima vemos a porcentagem de venda por cada Categoria!

```python
# Gráfico de barras horizontal por categoria e subcategoria
plt.figure(figsize=(14, 12), dpi=300)

# Cores por categoria
categoria_cores = {'Furniture': '#5d00de', 'Office Supplies': '#0ee84f', 'Technology': '#e80e27'}
cores = [categoria_cores[cat] for cat in q10['Categoria']]

# Barras horizontais
bars = plt.barh(q10['SubCategoria'], q10['Valor_Venda'], color=cores, alpha=0.8, edgecolor='white', linewidth=1)

# Valores e porcentagens nas barras
total_vendas = sum(q10['Valor_Venda'])
for i, (bar, value) in enumerate(zip(bars, q10['Valor_Venda'])):
    percentage = (value/total_vendas*100)
    value_formatted = f"{value:,.0f}".replace(",", ".")
    plt.text(value + 5000, bar.get_y() + bar.get_height()/2, 
             f'${value_formatted} ({percentage:.1f}%)', 
             ha='left', va='center', fontweight='bold', fontsize=10)

# Labels e formatação
plt.title('Total de Vendas - Top 12 SubCategorias por Categoria', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Valor de Vendas ($)', fontsize=12, fontweight='bold')
plt.ylabel('SubCategorias', fontsize=12, fontweight='bold')

# Legenda das categorias
from matplotlib.patches import Patch
legend_elements = [Patch(facecolor=cor, label=cat) for cat, cor in categoria_cores.items()]
plt.legend(handles=legend_elements, title='Categorias', loc='center right', fontsize=10)

# Total no canto superior direito
total_formatted = f"{int(total_vendas):,}".replace(",", ".")
plt.text(0.98, 0.95, f'Total: ${total_formatted}', 
         transform=plt.gca().transAxes, ha='right', va='top', fontsize=12, fontweight='bold',
         bbox=dict(boxstyle='round', facecolor='lightgray', alpha=0.8))

# Ajustar layout
plt.tight_layout()
plt.show()
```

#### Assim temos o grafico de "Pizza" das categorias e suas subcategorias represetando a porcentagem de cada sua subcategoria
